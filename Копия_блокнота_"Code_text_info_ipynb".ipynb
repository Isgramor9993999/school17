###<font color='darkblue'>**Задание 3** (уровень 2)</font>

Составьте программу, аналогичную программе $Tabl\_code$, которая выводит двоичные коды символов, в соответствии с приведённым образцом.  


####**Образец форматированного вывода результатов работы**

01.JPG

####**Ваша программа:**

```python
kod = [i for i in range(33,123)] + [i for i in range(1040,1104)]

for i in range(len(kod)):
  if i % 10 == 0: print()
  f = '0' * (11-len(str(bin(i)[2:]))) + str(bin(i)[2:])
  print("{simv:>3} {k:>5}".format(simv =chr(kod[i]), k=f),end = '' )
```

###<font color='darkblue'>**Задание 4** (уровень 2)</font>

Составьте программу, аналогичную программе $Tabl\_code$, которая выводит 16-ричные коды символов, в соответствии с приведённым образцом.  


####**Образец форматированного вывода результатов работы**

02.JPG

####**Ваша программа:**

```python
kod = [i for i in range(33,123)] + [i for i in range(1040,1104)]

for i in range(len(kod)):
  if i % 10 == 0: print()
  f = '0' * (3-len(str(hex(kod[i])[2:]))) + str(hex(kod[i])[2:])
  print("{simv:>3} {k:>5}".format(simv =chr(kod[i]), k=f),end = '' )
```
